```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.express as px
```

```python
datos = yf.Ticker('CLP=X')
USD_2022 = datos.history(start="2022-01-01")
USD_2021 = datos.history(start="2021-01-01", end="2021-12-31")
USD_2022
```

```python
fig = px.line(USD_2022['Close'])
fig.show()
```

Two Y Axes

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig.data = []

fig.add_trace(go.Scatter(x=USD_2022.index, y=USD_2022["Close"], name="2022"))
fig.add_trace(go.Scatter(x=USD_2022.index, y=USD_2021["Close"], name="2021"))

# Set x-axis title
fig.update_xaxes(title_text="Date")

# Set x-axis title
fig.update_yaxes(title_text="USD Price in <b>CLP</b>")


display(fig)
```
