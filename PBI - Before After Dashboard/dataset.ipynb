```python
## Librerias necesarias
import pandas as pd
import random
from datetime import datetime, timedelta
from faker import Faker
```

```python
# Rango de fechas para las ventas
start_date = datetime(2024, 1, 1)
end_date = datetime(2025, 6, 18)

# Diccionario de productos con su ID, nombre de modelo y categoría
product_catalog = {
    "SUP-001": ("Essentials Oversized Tee", "Superior"),
    "SUP-002": ("Skateboarding Logo Hoodie", "Superior"),
    "SUP-003": ("Graphic Crewneck “Astro Drift”", "Superior"),
    "SUP-004": ("Classic Polo Black", "Superior"),
    "SUP-005": ("Denim Jacket Vintage Wash", "Superior"),
    "INF-006": ("Cargo Pants Tactical Fit", "Inferior"),
    "INF-007": ("Ripped Skinny Jeans 'No Mercy'", "Inferior"),
    "INF-008": ("Sweatpants Classic Grey", "Inferior"),
    "INF-009": ("Track Pants “Streets”", "Inferior"),
    "INF-010": ("Biker Shorts Urban Black", "Inferior"),
    "CAL-011": ("Air Force 1 ‘07 Triple White", "Calzado"),
    "CAL-012": ("Yeezy Boost 350 V2 Zebra", "Calzado"),
    "CAL-013": ("Old Skool Black/White", "Calzado"),
    "CAL-014": ("Chuck 70 Hi Canvas Natural", "Calzado"),
    "CAL-015": ("RS-X Efekt Gradient", "Calzado"),
    "ACC-016": ("Bucket Hat “Street Jungle”", "Accesorios"),
    "ACC-017": ("Crossbody Bag Mini Messenger", "Accesorios"),
    "ACC-018": ("Gafas de sol “Neo Matrix”", "Accesorios"),
    "ACC-019": ("Cinturón de lona con hebilla", "Accesorios"),
    "ACC-020": ("Cadena plateada Miami Cuban Link", "Accesorios")
}

# Lista de IDs de producto
product_ids = list(product_catalog.keys())

# Canales de venta y clientes ficticios
channels = ["Tienda Física", "Instagram", "Tienda Online", "TikTok Shop"]
clients = [f"Cliente_{i}" for i in range(1, 201)]

# Rango de fechas a simular
dates = pd.date_range(start=start_date, end=end_date, freq='D')

# Generador de ventas
sales_data = []
for date in dates:
    for _ in range(random.randint(1, 8)):  # ventas por día
        product_id = random.choice(product_ids)
        modelo, categoría = product_catalog[product_id]
        unidades = random.randint(1, 5)
        precio_unitario = random.randint(15000, 45000)
        monto_total = unidades * precio_unitario
        sales_data.append({
            "fecha_venta": date,
            "id_producto": product_id,
            "modelo": modelo,
            "categoría": categoría,
            "canal_venta": random.choice(channels),
            "cliente": random.choice(clients),
            "unidades": unidades,
            "precio_unitario": precio_unitario,
            "monto_total": monto_total
        })

# Crear DataFrame
df_ventas = pd.DataFrame(sales_data)

# Guardar a CSV
df_ventas.to_csv('./dataset_ventas.csv', index=False)
```

```python
fake = Faker('es_CL')
# Generar datos de clientes
clientes_data = []
for i in range(1, 201):
    id_cliente = f"cliente_{i}"
    nombre = fake.name()
    email = f"{nombre.replace(' ', '.').lower()}.{id_cliente.split('_')[1]}@email.com"
    telefono = fake.phone_number()
    ciudad = random.choice([
        "Santiago", "Valparaíso", "Concepción", "La Serena",
        "Antofagasta", "Temuco", "Rancagua", "Talca", "Iquique", "Puerto Montt"
    ])
    tipo_cliente = random.choice(["premium", "regular", "nuevo"])
    clientes_data.append({
        "id_cliente": id_cliente,
        "nombre": nombre,
        "email": email,
        "telefono": telefono,
        "ciudad": ciudad,
        "tipo_cliente": tipo_cliente
    })
# Crear DataFrame de clientes
df_clientes = pd.DataFrame(clientes_data)
# Guardar el DataFrame de clientes en un archivo CSV
df_clientes.to_csv('./dataset_clientes.csv', index=False)
```
